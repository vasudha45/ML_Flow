```python
import mlflow
import mlflow.pytorch
import torch
from pathlib import Path
import pandas as pd

# Load the results.csv file
results = pd.read_csv('runs/train/exp/results.csv')
results.columns = results.columns.str.strip()  # Strip leading and trailing spaces

# Get the last row of results (last epoch)
last_epoch = results.iloc[-1]

# Extract the metrics
precision = last_epoch['metrics/precision']
recall = last_epoch['metrics/recall']
map_05 = last_epoch['metrics/mAP_0.5']

# Set up MLflow tracking
mlflow.set_experiment("YOLOv5 COCO128 Experiment")

# Define the YOLOv5 model path
model_path = Path("runs/train/exp/weights/best.pt")

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', model_path)

with mlflow.start_run():
    # Log parameters
    mlflow.log_param("epochs", 5)
    mlflow.log_param("model", "yolov5s")

    # Log the trained model weights
    mlflow.pytorch.log_model(
        pytorch_model=model,
        artifact_path="models"
    )

    # Log metrics
    mlflow.log_metric("precision", precision)
    mlflow.log_metric("recall", recall)
    mlflow.log_metric("mAP_0.5", map_05)  # Updated metric name

    # Register the model
    mlflow.register_model("yolov5_coco128_model", "models")

print("Logged metrics and model to MLflow:")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"mAP_0.5: {map_05}")
```
